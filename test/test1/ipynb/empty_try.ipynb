```python
import matplotlib.pyplot as plt
import numpy as np
import healpy as hp
%matplotlib inline
```

```python
# cool, viridis, jet
cmap = plt.cm.viridis
cmap.set_under("w")
```

```python
# fits and map settings

main_folder = '../../../data/'
name = 'HFI_SkyMap_143-field-IQU_2048_R2.02_full.fits'
mask = False
only_temp = True
nmax = 2048
lmax = 512
```

```python
if only_temp:
    cmbmap = hp.read_map(main_folder + name, field=(0))
else:
    if mask:
        cmbmap = hp.read_map(name, field=(0, 1, 2, 3, 4))
    else:
        cmbmap = hp.read_map(name, field=(0, 1, 2))
```

```python
# plot settings

plot_map = cmbmap
coord = 'G' # 'G', ['G', 'E'] or ['G', 'C']
xsize = 2000
title = ''
min_value = 0.0
max_value = 0.0
cbar = True
notext = False
norm = 'hist' # None, 'hist' or 'log'

flag_out = False
name_out = name + '-P1.png'
```

```python
# mollview, cartview, orthview

if min_value != 0.0 or max_value != 0.0:
    plot_map_out = hp.mollview(plot_map, coord=coord, xsize=xsize, title=title, 
                               min=min_value, max=max_value, 
                               cbar=cbar, notext=notext, norm=norm, cmap=cmap, return_projected_map=True)
else:
    plot_map_out = hp.mollview(plot_map, coord=coord, xsize=xsize, title=title, 
                               cbar=cbar, notext=notext, norm=norm, cmap=cmap, return_projected_map=True)

if flag_out:
    plt.savefig(name_out, dpi=400)
```
